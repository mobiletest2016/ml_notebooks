```python
import tensorflow
```

```python
from tensorflow import keras
```

```python
from keras.datasets import fashion_mnist
```

```python
(train_X, train_y), (test_X, test_y) = fashion_mnist.load_data()
```

```python
train_X.shape
```

```python
test_X.shape
```

```python
train_y.shape
```

```python
test_y.shape
```

```python
train_X = train_X / 255.0
test_X = test_X / 255.0
```

```python
import matplotlib.pyplot as plt
```

```python
from keras import layers, models, losses, optimizers, metrics
```

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(512, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer=optimizers.AdamW(learning_rate=0.001, weight_decay=0.001), loss=losses.SparseCategoricalCrossentropy(), metrics=metrics.SparseCategoricalAccuracy())
```

```python
model.fit(train_X, train_y, epochs=10)
```

```python
model.evaluate(test_X, test_y)
```

```python
model_cnn = models.Sequential([])
```

```python
train_X = train_X.reshape((-1, 28, 28, 1))
test_X = test_X.reshape((-1, 28, 28, 1))
```

```python
model_cnn = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model_cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model_cnn.fit(train_X, train_y, epochs=10)
```

```python
model_cnn.evaluate(test_X, test_y)
```

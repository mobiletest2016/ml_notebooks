```python
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=150, factor=0.4, noise=0.3, random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:, 1], c=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=50)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
plt.figure(figsize=(15,25))
for i,samples in enumerate((1,5,10,15,20,25)):
    clf = DecisionTreeClassifier(min_samples_leaf=samples)
    clf.fit(X_train,y_train)
    ax = plt.subplot(4, 2, i + 1)
    DecisionBoundaryDisplay.from_estimator(
            clf, X_test, response_method="predict",
            ax=ax, alpha=0.5,)
    ax.scatter(X[:,0], X[:, 1], c=y)
    ax.text(-1.8,1.8,"Min Samples: " + str(samples) + ", Score: " + str(clf.score(X_test, y_test))[:4])

plt.show()
```
